```python
import pandas as pd

price_df = pd.read_csv('prices.csv')

price_df
```

```python
price_df = pd.read_csv('prices.csv', names=['ticker', 'date', 'open','high', 'low',
                                            'close', 'volume','adj_close','adj_volume'])

price_df
```

```python
price_df.median(axis=0)
```
